```python
# segregating depending upon mean mode median or just None to fill values
for i in ["MasVnrType","Electrical"]:
    df[i].fillna(value=df[i].mode()[0], inplace = True)
```

```python
df.isna().sum()[df.isna().any()]
```

```python
# filling with mean(for continuous variables)

for j in ["LotFrontage", "MasVnrArea"]:
    df[j].fillna(value=df[j].mean(), inplace=True)
```

```python
df.isna().sum()[df.isna().any()]
```

```python
# Filling with None for none valued Nans

for k in ["BsmtQual",
"BsmtCond",
"BsmtExposure",
"BsmtFinType1",
"BsmtFinType2",
"FireplaceQu",
"GarageType",
"GarageYrBlt",
"GarageFinish",
"GarageQual",
"GarageCond"]:
    df[k].fillna(value="None", inplace= True)
```
