### Lets import the libraries

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

### Lets get the Data

```python
data = pd.read_csv("Churn_Modelling.csv")
df = data.copy()
df
```

### Data Pre-processing

```python
x = df.iloc[:,3:-1]
y = df.iloc[:,-1:]
```

```python
con_cols = [col for col in x.columns if x[col].dtype != 'object']
cat_cols = [col for col in x.columns if x[col].dtype == 'object']
```

```python
con_cols
```

```python
cat_cols
```

```python
for i in ["HasCrCard", "IsActiveMember"]:
    con_cols.remove(i)

con_cols
```

### Lets split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2 ,random_state=42)
```

### Encoding and Scalling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
sc = StandardScaler()
oe = OneHotEncoder(handle_unknown= "ignore")
```

```python
con_cols
```

```python
x_train[con_cols] = sc.fit_transform(x_train[con_cols])
x_test[con_cols] = sc.transform(x_test[con_cols])
```

```python
x_train
```

```python
x_test
```

```python
oe_train = pd.DataFrame((oe.fit_transform(x_train[cat_cols])).toarray(),index = x_train.index)
oe_test = pd.DataFrame((oe.transform(x_test[cat_cols])).toarray(), index = x_test.index)
```

```python
oe_test
```

```python
oe_train
```

```python
for i in ["HasCrCard", "IsActiveMember"]:
    con_cols.append(i)

con_cols
```

```python
train_final = pd.concat([x_train[con_cols],oe_train], axis= 1)
test_final = pd.concat([x_test[con_cols],oe_test], axis= 1)
```
